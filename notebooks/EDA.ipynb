```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the CSV files
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')

# Check the first few rows to understand the structure of your data
print(train.head())
print(test.head())

# Replace 'promotion' with the actual column name representing promotions in your data
promotion_column = 'Promo'  # Update this if the column name is different

# Calculate distribution percentages
train_dist = train[promotion_column].value_counts(normalize=True) * 100
test_dist = test[promotion_column].value_counts(normalize=True) * 100

# Combine distributions into a single DataFrame for comparison
distribution_df = pd.DataFrame({'Train': train_dist, 'Test': test_dist}).reset_index()
distribution_df.columns = ['Promotion', 'Train Distribution (%)', 'Test Distribution (%)']

# Print distributions
print(distribution_df)

# Visualization
plt.figure(figsize=(10, 6))
sns.barplot(data=pd.melt(distribution_df, id_vars='Promotion', 
value_vars=['Train Distribution (%)', 'Test Distribution (%)']),
x='Promotion', y='value', hue='variable')
plt.title('Promotion Distribution in Train vs Test')
plt.ylabel('Distribution (%)')
plt.xlabel('Promotion')
plt.show()
```
